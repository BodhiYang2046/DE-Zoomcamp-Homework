```python
import pandas as pd
from time import time
from sqlalchemy import create_engine
```

```python
pd.__version__
```

```python
df = pd.read_csv('green_tripdata_2019-09.csv', low_memory=False, nrows=100)
```

```python
print(pd.io.sql.get_schema(df, name='green_tripdata'))
```

```python
pd.to_datetime(df.lpep_pickup_datetime)
```

```python
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
print(pd.io.sql.get_schema(df, name='green_tripdata', con=engine))
```

```python
df_iter = pd.read_csv('green_tripdata_2019-09.csv', iterator=True, low_memory=False, chunksize=100000)
```

```python
df = next(df_iter)
```

```python
df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
```

```python
df.head()
```

```python
%time

df.to_sql(name='green_tripdata_trip', con=engine, index=False, if_exists='replace')
```

```python
%time
df.head(0).to_sql(name='green_tripdata_trip', con=engine, index=False, if_exists='replace')
```

```python
while True:
    t_s = time()
    df = next(df_iter)
    
    df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
    df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
    df.to_sql(name='green_tripdata_trip', con=engine, index=False, if_exists='append')
    
    t_e = time()
    
    print('inserted another chunk, took %.3f second'%(t_e - t_s))
```

```python
query = """
SELECT * FROM green_tripdata_trip LIMIT 10
"""

pd.read_sql(query, con=engine)
```
